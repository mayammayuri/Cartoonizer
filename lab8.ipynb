```python
# -*- coding: utf-8 -*-
"""
Created on Wed Oct 14 11:18:32 2020

@author: mayuri mane
"""


import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from skimage import io,data
from skimage.color import rgb2gray
from skimage import data, img_as_float
from skimage.restoration import denoise_tv_chambolle, denoise_bilateral
from PIL import Image
import timeit


def bilateral(image):

    noisy = img_as_float(image)

    fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(8, 5))

    plt.gray()

    ax[0, 0].imshow(noisy)
    ax[0, 0].set_title('noisy')
    ax[0, 2].imshow(denoise_bilateral(noisy, sigma_spatial=1))
    ax[0, 2].set_title('Bilateral')

    ax[1, 1].imshow(denoise_bilateral(noisy, sigma_spatial=1))
    ax[1, 1].set_title('(more) Bilateral')
    ax[1, 2].imshow(noisy)
    ax[1, 2].set_title('original')

    fig.subplots_adjust(wspace=0.02, hspace=0.2,
                        top=0.9, bottom=0.05, left=0, right=1)

    plt.show()
    return (denoise_bilateral(noisy, sigma_spatial=1))

start = timeit.default_timer()
img=io.imread("noisy_img1.jpg")
#img=data.chelsea()
img=rgb2gray(img)
img_color=bilateral(img)
stop = timeit.default_timer()
print('Time: ', stop - start,"secs")  
img_gray=rgb2gray(img)
plt.figure(1)
plt.subplot(1,2,1)
plt.imshow(img_gray)
plt.subplot(1,2,2)
plt.imshow(img_color)
plt.show()

```
